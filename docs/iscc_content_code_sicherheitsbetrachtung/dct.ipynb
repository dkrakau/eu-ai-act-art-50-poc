```python
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
import numpy as np
```

```python
# Definitionsbereich
x_max = 16
x = np.linspace(0, x_max, 500)  # 500 Punkte zwischen 0 und 8
```

```python
# Funktionen ohne Verschiebung
f1 = 2 * np.cos(x * np.pi / 32)
f2 = 2 * np.cos(x * np.pi / 16)
f3 = 2 * np.cos(x * np.pi / 8)
f4 = 2 * np.cos(x * np.pi / 4)
f5 = 2 * np.cos(x * np.pi / 2)
f6 = 2 * np.cos(x * np.pi / 1)
```

```python
# Funktionen mit Verschiebung
g1 = 2 * np.cos((x + (1/2)) * np.pi / 32)
g2 = 2 * np.cos((x + (1/2)) * np.pi / 16)
g3 = 2 * np.cos((x + (1/2)) * np.pi / 8)
g4 = 2 * np.cos((x + (1/2)) * np.pi / 4)
g5 = 2 * np.cos((x + (1/2)) * np.pi / 2)
g6 = 2 * np.cos((x + (1/2)) * np.pi / 1)
```

```python
# Plot ohne Verschiebung
plt.figure(figsize=(15, 5))
plt.plot(x, f1, label=r"$2\cos\left(\frac{\pi}{32}x\right)$")
plt.plot(x, f2, label=r"$2\cos\left(\frac{\pi}{16}x\right)$")
plt.plot(x, f3, label=r"$2\cos\left(\frac{\pi}{8}x\right)$")
plt.plot(x, f4, label=r"$2\cos\left(\frac{\pi}{4}x\right)$")
plt.plot(x, f5, label=r"$2\cos\left(\frac{\pi}{2}x\right)$")
#plt.plot(x, f6, label=r"$2\cos\left(\frac{\pi}{1}x\right)$")
for i in range(x_max+1):
    plt.axvline(i, color="grey", linewidth=1, linestyle="--")
plt.xlabel("x")
plt.ylabel("y", rotation=0)
plt.title("DCT-Funktion ohne Verschiebung")
plt.grid(True)
plt.gca().xaxis.set_major_locator(MultipleLocator(1))  # 1er Schritte auf der x-Achse
plt.legend()
plt.show()
```

```python
# Plot mit Verschiebung
plt.figure(figsize=(15, 5))
plt.plot(x, g1, label=r"$2 \cos\left(\frac{\pi}{32}(x + \frac{1}{2})\right)$")
plt.plot(x, g2, label=r"$2 \cos\left(\frac{\pi}{16}(x + \frac{1}{2})\right)$")
plt.plot(x, g3, label=r"$2 \cos\left(\frac{\pi}{8}(x + \frac{1}{2})\right)$")
plt.plot(x, g4, label=r"$2 \cos\left(\frac{\pi}{4}(x + \frac{1}{2})\right)$")
plt.plot(x, g5, label=r"$2 \cos\left(\frac{\pi}{2}(x + \frac{1}{2})\right)$")
#plt.plot(x, g6, label=r"$2 \cos\left(\frac{\pi}{1}(x + \frac{1}{2})\right)$")
for i in range(x_max+1):
    plt.axvline(i, color="grey", linewidth=1, linestyle="--")
plt.xlabel("x")
plt.ylabel("y", rotation=0)
plt.title("DCT-Funktion mit Verschiebung")
plt.grid(True)
plt.gca().xaxis.set_major_locator(MultipleLocator(1))  # 1er Schritte auf der x-Achse
plt.legend(loc="lower left")
plt.show()
```
