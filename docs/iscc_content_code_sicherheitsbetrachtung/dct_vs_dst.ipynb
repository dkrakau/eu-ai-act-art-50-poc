```python
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
import numpy as np
```

```python
# Daten vorbereiten
x = np.linspace(-np.pi, np.pi, 500)
y_sin = np.sin(x)

# Punkte bei ±pi/3 berechnen
x_points = [np.pi / 3, -np.pi / 3]
y_sin_points = [np.sin(v) for v in x_points]

# Plot erstellen
fig, ax = plt.subplots(figsize=(8, 5))

# Sinuskurve
ax.plot(x, y_sin, color='orange', linewidth=2, label='sin(x)')
ax.scatter(x_points, y_sin_points, color='black', s=60, zorder=5)
ax.plot(x_points, y_sin_points, color='gray', linestyle='--', linewidth=1.5, alpha=0.8)

# Beschriftungen
ax.text(x_points[0] + 0.1, y_sin_points[0], f"sin(π·1/3) = {y_sin_points[0]:.2f}", color='black', fontsize=10, va='bottom')
ax.text(x_points[1] - 1.2, y_sin_points[1], f"sin(-π·1/3) = {y_sin_points[1]:.2f}", color='black', fontsize=10, va='bottom')

# Achsen-Einstellungen
ax.set_xlim(-np.pi + 0.2, np.pi + 0.2)
ax.set_ylim(-1.5, 1.5)
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# Achsen-Ticks und Labels
ax.set_yticks([-1, 1])
ax.set_yticklabels(['-1', '1'])
ax.set_xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
ax.set_xticklabels(['-π', r'-π·1/2', '0', r'π·1/2', 'π'])

# Achsenpfeile für beide Richtungen hinzufügen
arrow_style = dict(arrowstyle="->", color='black', linewidth=1.5)

# X-Achse Pfeile (nach rechts und links)
ax.annotate('', xy=(np.pi, 0), xytext=(0, 0), arrowprops=arrow_style)
ax.annotate('', xy=(-np.pi, 0), xytext=(0, 0), arrowprops=arrow_style)

# Y-Achse Pfeile (nach oben und unten)
ax.annotate('', xy=(0, 1.5), xytext=(0, 0), arrowprops=arrow_style)
ax.annotate('', xy=(0, -1.5), xytext=(0, 0), arrowprops=arrow_style)

# Achsenbeschriftungen
ax.text(np.pi + 0.2, 0, 'x', fontsize=12, va='center')
ax.text(0.1, 1.45, 'y', fontsize=12, ha='left')

# Gitter, Legende, Titel
ax.grid(True, linestyle=':', linewidth=0.7)
ax.legend(loc='upper right')
ax.set_title("Sinus-Symmetrie", fontsize=12)

plt.show()
```
